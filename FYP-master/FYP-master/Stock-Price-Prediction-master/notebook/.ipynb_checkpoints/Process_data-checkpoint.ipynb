```python
import pandas as pd
import word2vec as w2v
import numpy as np
import os
import sys
import math
import textblob
```

```python
data_path = os.path.join(os.path.dirname(os.getcwd()), 'data')
```

```python
os.listdir(data_path)
```

```python
stock_data_file_path = "stock_data_AAPL_2019-03-15_2019-04-15.csv"
stock_data = pd.read_csv(os.path.join(data_path, stock_data_file_path))

twitter_data_file_path = "twitter_data_english_Apple_2019-03-15_2019-04-15.csv"
twitter_data = pd.read_csv(os.path.join(data_path, twitter_data_file_path))
```

```python
stock_data['Date'] = pd.to_datetime(stock_data['Date'])
```

```python
def clean_twitter_data():
    twitter_data.text = twitter_data.text.str.replace('#', '')

def get_twitter_sentiment(value):
    
    def get_sentiment(value):
        textblob_obj = textblob.TextBlob(value)
        return textblob_obj.sentiment[1]
    
    twitter_data['sentiment'] = twitter_data['text'].apply(get_sentiment)


clean_twitter_data()
get_twitter_sentiment(twitter_data)
twitter_data_daywise = twitter_data.set_index(pd.to_datetime(twitter_data['date'])).resample('D').mean()
twitter_data_daywise['Date'] = twitter_data_daywise.index
stock_sentiments = stock_data.merge(twitter_data_daywise)
stock_sentiments.sentiment.fillna(stock_sentiments.sentiment.mean(), inplace=True)
```
